## Part 2_Experiment and metrics design

Q1.What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why would you choose this metric?

**A1: A simple but key metric of the experiment is the number of visits by Gotham' drivers to Metropolis and the number of visits by Metropolis' drivers to Gotham. Number of visits can straightforwardly reflect if driver partners serve both cities for more times, which is what the Ultimate managers care about mostly, at least at the first stage of the experiment; it is strongly related to revenue or profit; it is easy to measure.**

Q2.Describe a practical experiment you would design to compare the effectiveness of the proposed change in relation to the key measure of success. Please provide details on:

Q2_a. how you will implement the experiment

**A2_a: We can design and implement a A/B testing**
* **Firstly, develop a hypothesis. In this case, the null hypothesis is the observed raise of number of visit is due to the chance; the alternative hypothesis is the observed raise of number of visit is due to the police of reimbursement. We need to think about what size of impact is meaningful to detect, i.e., sensitivity, at this stage.** 
* **Secondly, randomly assign users to two groups. We can calculate needed sample size based on baseline number of visit, desired confidence level, desired power, and sensitivity that is chosen by us.**
* **Thirdly, expose one group to the current rule, i.e., no reimbursement, which is control group, and expose another group to the new rule, i.e., reimbursing toll cost, which is treatment group.**
* **Fourthly, evaluating the result by comparing the two group's performance.**
    
Q2_b. what statistical test(s) you will conduct to verify the significance of the observation

**A2_b: We can do z-test or t-test based on the sample size and population distribution. There is a permutation test designed as followed, which has no requirement of making any assumptions about the sampling distribution. Besides the overall performance, we'd better identify the difference between two city's divers, weekday and weekend, day and night, and so on. So if we decide to do A/B testing on multiple groups, we need to use ANOVA tests.**

Q2_c. how you would interpret the results and provide recommendations to the city operations team along with any caveats.

**A2_c: According to result(dummy experiment as followed), the raise of number of daily visit to another city result from the police of reimbursement. We need to provide more detailed information related to the result such as, confidence interval, sample size, desired power and so on. Alert: 1. The metric used here is based on the number of visit to another city, we need more monetary information to evaluate revenue or profit. 2. The result can not confirm the long-term impact of the new police.**

The following A/B testing is designed based on the Part 1 dataset. We assume the login time in the dataset is the time for a driver to pay the toll through the bridge. We use the daily count of the weekday in the last 28 days in the dataset as baseline, i.e., control group. Treatment group is the data of control group plus a normally distributed lift. This a one side test.

````python
# Define control group
con = weekday.tail(28).daily_count.to_frame().reset_index(drop=True)
con['group'] = 'C'
print(con.describe())
con.daily_count.hist()
````

````python
# Statistics of control group 
con_mean = con.daily_count.mean()
con_std = con.daily_count.std()
# Set sensitivity as 20%
lift_mean = con_mean * 0.2
lift_std = con_std * 0.2
````

````python
# Assume the lift of daily count to ditribute randomly
lift = pd.DataFrame(stats.norm.rvs(lift_mean, lift_std, 28).reshape(-1, 1),columns = ['daily_count'])
# Define treatment group as control group plus li
tre = con + lift
tre['group'] = 'T'
# Build a dataframe to store both control group data and treatment group data
test = pd.concat([con, tre], ignore_index=True)
````

````python
# Implement of a permutation test
difference = []
for _ in range(10000):
    test['permutational'] = np.random.permutation(test.daily_count)
    test_diff = test[test.group=='T'].permutational.mean() - test[test.group=='C'].permutational.mean()
    difference.append(test_diff)
````

````python
# Calculate p-value
p_value = 1 - stats.percentileofscore(difference, lift_mean) * 0.01
print('P value is {:.3f}'.format(p_value))
# Visualize
sns.histplot(difference, kde=True)
plt.axvline(lift_mean, color='g', lw=2)
plt.text(190,400,'lift_mean')
````

**We should have defined confidence level (0.1, 0.05, or 0.01) beforehand, but here the result is statistically significant since the p-value is less than the smallest confidence level.**

#### Missing Values

````python
def missing_vals(data_):
    # Check the number and percentage of missing values
    miss_     = data_.isnull().sum()
    miss_pct  = data_.isnull().sum()/data_.shape[0]
    
    miss_pct  = pd.concat([miss_, miss_pct], axis =1)
    miss_pct.reset_index(inplace=True)
    miss_cols = miss_pct.rename(columns={'index':'Column Name', 0:'Missings', 1:'Missing_pct'})
    
    miss_cols = miss_cols[miss_cols.iloc[:,1]!=0].sort_values('Missing_pct', ascending=False).round(3)
    miss_cols.reset_index(inplace=True, drop=True)
    
    return miss_cols

missing_retention = missing_vals(retention)
missing_retention
````

It's hard to display the strong correlation between the two features because of the extreme outliers. So we will remove the extreme outliers in both features to check the correlation

````python
def extreme_outlier(columns):
# Calculate the extreme outlier
    Q1, Q3 = retention[columns].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    ext_outlier = Q3 + 3 * IQR    
    print('For the feature of {}, the data with a value greater than {:.2f} can be considered as extreme outliers.'.
          format(columns,ext_outlier))
    return ext_outlier

# Check the four numeric features 
ext_outlier_avg_surge = extreme_outlier('avg_surge')
ext_outlier_surge_pct = extreme_outlier('surge_pct')
````

````python
# Keep the entries without extreme ouliers in avg_surge and surge_pct
avg_pct_surge = retention[(retention.avg_surge <= ext_outlier_avg_surge) & 
                          (retention.surge_pct <= ext_outlier_surge_pct)]

# Check the correlation between the two features
print(np.corrcoef(avg_pct_surge.avg_surge, avg_pct_surge.surge_pct))
_ = sns.jointplot(x='avg_surge', y='surge_pct', data=avg_pct_surge, kind='reg')
````

#### Defining Functions

````python
def model_fit_evaluate(clf):
    #Fit the classifier on the data
    clf.fit(X_train, y_train)
    
    #Predict
    predictions = clf.predict(X_test)
    
    # Metrics
    accuracy = accuracy_score(y_test, predictions).round(3)
    precision = precision_score(y_test, predictions).round(3)
    recall = recall_score(y_test, predictions).round(3)
    f1 = f1_score(y_test, predictions).round(3)
    print('Accuracy score is:', accuracy)
    print('Precision score is:', precision) 
    print('Recall score is:', recall)
    print('F1 score is:', f1)
    
    # Calculate the cross validation roc_auc score
    # Train set
    cv_score_tr= cross_val_score(clf,X_train,y_train,cv=5,scoring='roc_auc')
    print('CV ROC-AUC Score(Train): Mean - {:5f} | Std - {:5f} | Max - {:5f} | Min - {:5f}'.format(
        np.mean(cv_score_tr),np.std(cv_score_tr),np.max(cv_score_tr),np.min(cv_score_tr)))
    # Test set
    cv_score_te= cross_val_score(clf,X_test,y_test,cv=5,scoring='roc_auc')
    print('CV ROC-AUC Score(Test): Mean - {:5f} | Std - {:5f} | Max - {:5f} | Min - {:5f}'.format(
        np.mean(cv_score_te),np.std(cv_score_te),np.max(cv_score_te),np.min(cv_score_te)))
    
    # Plot ROC Curve
    fpr, tpr, thresholds = roc_curve(y_test, clf.predict_proba(X_test)[:, 1])
    plt.plot(fpr, tpr, linewidth=2)
    plt.plot([0, 1], [0, 1], 'r', linewidth=2)
    plt.axis([0, 1, 0, 1])
    plt.xlabel('False Positive Rate (FPR)', fontsize=15)
    plt.ylabel('True Positive Rate (TPR)', fontsize=15)
    plt.title('ROC Curve',fontsize=15)
    plt.xlim(-0.05,1)
    plt.ylim(0,1.05)
    plt.show()

    return accuracy, precision, recall, f1, cv_score_tr.mean().round(3), cv_score_te.mean().round(3)
````

````python
def grid_search(estimator, hyper_para):
    # Tuning hyper-parameters with Grid Search
    grid = GridSearchCV(estimator, param_grid=hyper_para, scoring='roc_auc', n_jobs=-1, cv=5, verbose=2)
    grid.fit(X_train, y_train)
    best_params = grid.best_params_
    best_score = grid.best_score_
    best_estimator = grid.best_estimator_
    print('Best Parameters:', best_params)
    print('Best roc_auc Score:', round(best_score, 3))
    print('Best Estimator:', best_estimator)
````

````python
def random_search(estimator, hyper_para, number_of_iter, number_of_folds):
    # Tuning hyper-parameters with Random Search
    random = RandomizedSearchCV(estimator, param_distributions=hyper_para, scoring='roc_auc', n_jobs =-1,
                                n_iter=number_of_iter, cv = number_of_folds, random_state=42,  verbose=2)
    random.fit(X_train, y_train)
    best_params = random.best_params_
    best_score = random.best_score_
    best_estimator = random.best_estimator_
    print('Best Parameters:', best_params)
    print('Best roc_auc Score:', round(best_score, 3))
    print('Best Estimator:', best_estimator)
````
